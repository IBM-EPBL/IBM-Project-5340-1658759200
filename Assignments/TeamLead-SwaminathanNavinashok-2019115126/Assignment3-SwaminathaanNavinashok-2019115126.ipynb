## 6. Find the outliers and replacing them

```python
outliers = data.quantile(q = (0.25, 0.75))
outliers
```

```python
a = data.Age.quantile(0.25)
b = data.Age.quantile(0.75)
c = b - a
lower_limit = a - 1.5 * c
data.median(numeric_only=True)
```

```python
data['Age'] = np.where(data['Age'] < lower_limit, 7, data['Age'])
sns.boxplot(x = data.Age, showfliers = False)
```

## 7. Checking for Categorical columns and performing encoding


```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

lab = LabelEncoder()
data.Sex = lab.fit_transform(data.Sex)

data.head()
```

## 8. Splitting the data into dependent and independent variables

```python
y = data["Sex"]
y.head()
```

```python
x = data.drop(columns=["Sex"], axis=1)
x.head()
```

## 9. Scale the independent variables

```python
from sklearn.preprocessing import scale
X_Scaled = pd.DataFrame(scale(x), columns=x.columns)
X_Scaled.head()
```

## 10. Splitting the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X_Scaled, y, test_size=0.2, random_state=0)
```

## 11. Building the Model

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10,criterion='entropy')

model.fit(X_Train,Y_Train)
```

```python
y_predict_train = model.predict(X_Train)
y_predict = model.predict(X_Test)
```

## 12. Training the Model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print('Training accuracy: ',accuracy_score(Y_Train,y_predict_train))
```

## 13.Testing the Model

```python
print('Testing accuracy: ',accuracy_score(Y_Test,y_predict))
```

## 14. Measuring the performance using Metrics

```python
pd.crosstab(Y_Test,y_predict)
```

```python
print(classification_report(Y_Test,y_predict))
```
